```python
for h in [2**i for i in range(3, 7)]:
    df1 = df[df['hid_feature']==h]
    df1['ratio'].plot()
```

```python
import matplotlib.pyplot as plt

# 建立畫布
plt.figure(figsize=(10, 5))

# 初始化 x 軸起點
start_idx = 0

# 按順序畫每段
for h in [2**i for i in range(3, 7)]:
    df1 = df[df['hid_feature']==h]
    y = df1['ratio'].values
    x = range(start_idx, start_idx + len(y))
    
    plt.plot(x, y, label=f'hid={h}')
    start_idx += len(y)

plt.xlabel('Concatenated index')
plt.ylabel('hid_feature')
plt.title('Ratio by hid_feature segments (with different colors)')
plt.legend()
plt.show()
```
